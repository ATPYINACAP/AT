# **ACTIVIDAD 1:**

Considerando la base de datos “Perfil Lipídico.xlsx”. Determinar la media, mediana y moda de las variables.

```python
df.columns
```

```python
import statistics

# Calcular la cantidad de modas usando statistics.multimode()
cant_modas1 = len(statistics.multimode(df['Colesterol Total']))
cant_modas2 = len(statistics.multimode(df['Colesterol HDL']))
cant_modas3 = len(statistics.multimode(df['Colesterol VLDL']))
cant_modas4 = len(statistics.multimode(df['Colesterol LDL']))

# Imprimir la cantidad de modas por variable
print("La cantidad de modas por variable:\n",
      'Colesterol Total:', cant_modas1, "\n",
      'Colesterol HDL:', cant_modas2, "\n",
      'Colesterol VLDL:', cant_modas3, "\n",
      'Colesterol LDL:', cant_modas4)
```

```python
# Crear el DataFrame con la media, mediana y moda de 'Colesterol Total'
d1 = pd.DataFrame({
    'Colesterol Total': [
        round(df['Colesterol Total'].mean(), 2),
        round(df['Colesterol Total'].median(), 2),
        statistics.multimode(df['Colesterol Total'])[0],
    ]
}, index=['Media', 'Mediana','Moda'])

# Imprimir el DataFrame
print(d1)
```

```python
statistics.multimode(df['Colesterol HDL'])
```

```python
# Crear el DataFrame con la media, mediana y moda de 'Colesterol HDL'
d2 = pd.DataFrame({
    'Colesterol HDL': [
        round(df['Colesterol HDL'].mean(), 2),
        round(df['Colesterol HDL'].median(), 2),
        statistics.multimode(df['Colesterol HDL'])[0],
        statistics.multimode(df['Colesterol HDL'])[1],
        statistics.multimode(df['Colesterol HDL'])[2],
        statistics.multimode(df['Colesterol HDL'])[3]
    ]
}, index=['Media', 'Mediana', 'Moda 1','Moda 2','Moda 3','Moda 4'])

# Imprimir el DataFrame
print(d2)
```

```python
# Crear el DataFrame con la media, mediana y moda de 'Colesterol VLDL'
d3 = pd.DataFrame({
    'Colesterol VLDL': [
        round(df['Colesterol VLDL'].mean(), 2),
        round(df['Colesterol VLDL'].median(), 2),
        statistics.multimode(df['Colesterol VLDL'])[0],
        statistics.multimode(df['Colesterol VLDL'])[1]
    ]
}, index=['Media', 'Mediana', 'Moda 1',"Moda 2"])

# Imprimir el DataFrame
print(d3)
```

```python
# Crear el DataFrame con la media, mediana y moda de 'Colesterol LDL'
d4 = pd.DataFrame({
    'Colesterol LDL': [
        round(df['Colesterol LDL'].mean(), 2),
        round(df['Colesterol LDL'].median(), 2),
        statistics.multimode(df['Colesterol LDL'])[0],
        statistics.multimode(df['Colesterol LDL'])[1]
    ]
}, index=['Media', 'Mediana', 'Moda 1',"Moda 2"])

# Imprimir el DataFrame
print(d4)
```

# **ACTIVIDAD 2:**

Obtener el Histograma, gráfico Cuantil-Cuantil y Boxplot para cada una de las variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np

# Definir el número de bins usando la regla de Sturges
n = len(df['Colesterol Total'])
bins = int(1 + math.log2(n))

# Definir el rango (máximo - mínimo) para calcular la amplitud de los intervalos
rango = max(df['Colesterol Total']) - min(df['Colesterol Total'])
amp = rango / bins

# Crear un histograma
plt.hist(df['Colesterol Total'], bins=bins, color='skyblue', edgecolor='black')

# Agregar etiquetas a los ejes con un tamaño de fuente mayor
plt.title('Histograma de Colesterol Total\n', fontsize=18)

plt.xlabel('\nColesterol Total (mg/dl)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# Ajustar límites de los ejes
plt.ylim(0, 35)
plt.xlim(0.95 * min(df['Colesterol Total']), max(df['Colesterol Total']) * 1.05)

# Establecer marcas en el eje X con el paso de 'amp'
plt.xticks(np.arange(round(min(df['Colesterol Total']), 1), round(max(df['Colesterol Total']) + 1, 1), amp))

# Activar la grilla
plt.grid(True, which='both', linestyle='-.', linewidth=0.5, color='gray')

# Mostrar el gráfico
plt.show()
```

```python
# Definir el número de bins usando la regla de Sturges
n = len(df['Colesterol Total'])
bins = int(1 + math.log2(n))

# Definir el rango (máximo - mínimo) para calcular la amplitud de los intervalos
rango = max(df['Colesterol Total']) - min(df['Colesterol Total'])
amp = rango / bins

# Crear el histograma y la curva de densidad
plt.hist(df['Colesterol Total'], bins=bins, color='skyblue', edgecolor='black', alpha=0.6, density=True)

# Agregar la curva de densidad (KDE)
sns.kdeplot(df['Colesterol Total'], color='purple', linewidth=2,linestyle='--')

# Agregar etiquetas a los ejes con un tamaño de fuente mayor
plt.title('Histograma de Colesterol Total con Densidad\n', fontsize=18)

plt.xlabel('\nColesterol Total (mg/dl)', fontsize=14)
plt.ylabel('Densidad', fontsize=14)  # Cambia a densidad, ya que ahora la altura de las barras representa densidad

# Ajustar límites de los ejes
plt.xlim(0.95 * min(df['Colesterol Total']), max(df['Colesterol Total']) * 1.05)

# Establecer marcas en el eje X con el paso de 'amp'
plt.xticks(np.arange(round(min(df['Colesterol Total']), 1), round(max(df['Colesterol Total']) + 1, 1), amp))

# Activar la grilla
plt.grid(True, which='both', linestyle='-.', linewidth=0.5, color='gray')

# Mostrar el gráfico
plt.show()
```

```python
#!pip install statsmodels==0.13.2
import statsmodels.api as sm

# Crear el gráfico Q-Q plot con statsmodels
qqplot = sm.qqplot(df['Colesterol Total'], line='s')

# Cambiar el color de los puntos
plt.gca().get_lines()[0].set_markerfacecolor('skyblue')  # Cambiar el color de los puntos
plt.gca().get_lines()[0].set_markeredgecolor('blue')  # Cambiar el borde de los puntos

# Cambiar el color y grosor de la línea de referencia
plt.gca().get_lines()[1].set_color('black')  # Cambiar el color de la línea
plt.gca().get_lines()[1].set_linewidth(3)  # Engrosar la línea

# Agregar título y etiquetas
plt.title('Gráfico Cuantil-Cuantil (Q-Q) para Colesterol Total\n', fontsize=18)
plt.xlabel('Cuantiles teóricos', fontsize=14)
plt.ylabel('Cuantiles de los datos', fontsize=14)

# Mostrar el gráfico
plt.grid(False, which='both', linestyle='-.', linewidth=0.5, color='gray')
plt.show()
```

```python
# Calcular el promedio (media) de 'Colesterol Total'
mean_value = np.mean(df['Colesterol Total'])

# Configurar el estilo de los outliers (flierprops),
# bigotes (whiskerprops), caja (boxprops), mediana (medianprops) y
# límites de los bigotes (capprops)
flierprops = dict(marker='o', markerfacecolor='red', markeredgecolor='red', markersize=8)
whiskerprops = dict(color='black', linewidth=2)
boxprops = dict(facecolor='skyblue', edgecolor='black', linewidth=2)
medianprops = dict(color='black', linewidth=2)
capprops = dict(color='black', linewidth=3)

# Crear el box plot para 'Colesterol Total'
plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura a 8x6 pulgadas
sns.boxplot(y=df['Colesterol Total'],
            flierprops=flierprops,
            whiskerprops=whiskerprops,
            boxprops=boxprops,
            medianprops=medianprops,
            capprops=capprops)

# Marcar el promedio en el gráfico
plt.scatter(y=mean_value, x=[0], color='blue', zorder=5, marker='X',s=100, label='Promedio')

# Agregar título y etiquetas para el gráfico
plt.title('Box Plot de Colesterol Total\n', fontsize=18)
# - title: Establece el título del gráfico, con un tamaño de fuente de 18.
plt.ylabel('Colesterol Total (mg/dl)', fontsize=14)
# - ylabel: Etiqueta el eje Y con "Colesterol Total (mg/dl)", con un tamaño de fuente de 14.

# Mostrar la grilla para facilitar la lectura del gráfico
plt.grid(False, which='both', linestyle='-.', linewidth=0.5, color='gray')

# Mostrar el gráfico
plt.show()  # Renderiza y muestra el gráfico en pantalla
```
