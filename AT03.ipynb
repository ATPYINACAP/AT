# **ACTIVIDAD 01:**
La compañía afirma que las personas no son capaces de detectar una diferencia de sabor entre los nuevos y los viejos aceites. Suponiendo que esta cifra es correcta (como proporción de largo plazo)


```python
# a) Determine la probabilidad de éxito, aquellos
#    clientes que detectan la diferencia

# Identificar las distintas categorias de una variables
df['Resultados'].unique()

# Determina el total de resultados
Total=df['Resultados'].count()

# Determina cantidad de valores de una categoria
No=df['Resultados'].value_counts()['No detecta una diferencia de sabor']

# Prob de exito
p_exito=No/Total

# Muestra la prob de éxito
print("La Probabilidad de éxito es: ",p_exito)
```

```python
# b) Determine la probabilidad de fracaso, de aquellos clientes
#   que detectan la diferencia

q_fracaso=1-p_exito
# Muestra la prob de fracaso
print("La Probabilidad de éxito es: ",round(q_fracaso,4))
```

```python
#c) Realice un gráfico de barras para representar la
#  probabilidad de éxito y fracaso
data = {'Probabilidad': [p_exito, round(q_fracaso,4)]}
dgraf = pd.DataFrame(data, index=['Éxito', 'Fracaso'])
```

```python
import matplotlib.pyplot as plt

dgraf.plot.bar(rot=0, color=['green', 'red'],width=0.25)
plt.ylabel("Probabilidad")
plt.title('Probabilidad de éxito y fracaso')

for i, v in enumerate(dgraf['Probabilidad']):
  plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# a) Usando la probabilidad de éxito de la actividad 01,
#    grafique la distribución binomial con n=10.

# Parámetros de la distribución binomial
n = 10  # Número de ensayos
p = 0.17  # Probabilidad de éxito
mu = n * p  # Media de la distribución normal aproximada
sigma = np.sqrt(n * p * (1 - p))  # Desviación estándar de la normal aproximada

# Crear el rango de valores para la variable binomial
x = np.arange(0, n + 1)

# Calcular las probabilidades de la distribución binomial
binom_probs = binom.pmf(x, n, p)

# Valores para la distribución normal aproximada
x_norm = np.linspace(0, n, 100)
norm_approx = norm.pdf(x_norm, mu, sigma)

# Crear el gráfico
plt.figure(figsize=(8, 6))

# Graficar la distribución binomial
plt.bar(x, binom_probs, color='pink', alpha=0.8, width=0.5,
        label=f'Distribución Binomial (n={n}, p={p})')

# Graficar la curva de densidad de la distribución normal aproximada
plt.plot(x_norm, norm_approx, 'r-', label='Aproximación Normal', lw=1,
         color='purple')

# Añadir títulos y etiquetas
plt.title("Distribución Binomial con Aproximación Normal", fontsize=14)
plt.xlabel("Número de éxitos", fontsize=12)
plt.ylabel("Probabilidad / Densidad", fontsize=12)
plt.legend()
plt.grid(True,linestyle='--')

# Mostrar el gráfico
plt.show()
```

# **ACTIVIDAD 03:**
Si consideramos una muestra aleatoria de 200 individuos que han comprado papas a la francesa en McByron:

```python
# a)
from scipy.stats import norm

# Parámetros del problema
n = 200  # Número de personas encuestadas
p = 0.17  # Probabilidad de éxito (notar la diferencia)
x = 25  # Valor para el cual queremos calcular la probabilidad

# Calcular la media y la desviación estándar de la distribución binomial
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Aplicar corrección por continuidad para la aproximación normal
z_score = (x - 0.5 - mu) / sigma

# Calcular la probabilidad usando la aproximación normal
probabilidad = 1 - norm.cdf(z_score)

print(f"La probabilidad de que al menos 180 personas noten la diferencia es: {probabilidad:.4f}")
```

```python
#b)
import numpy as np
from scipy.stats import norm

# Parámetros del problema
n = 200  # Número total de personas encuestadas
p = 0.83  # Probabilidad de que una persona no note la diferencia
x = 170  # Queremos calcular P(X <= 180)

# Calcular la media y la desviación estándar de la distribución binomial
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Aplicar corrección por continuidad y calcular el Z-score
z_score = (x + 0.5 - mu) / sigma

# Calcular la probabilidad usando la aproximación normal
probabilidad = norm.cdf(z_score)

# Mostrar el resultado
print(f"La probabilidad de que a lo más 170 personas no puedan notar la diferencia es: {probabilidad:.4f}")
```
